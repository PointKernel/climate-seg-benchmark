# Climate Segmentation Scaling Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.rcParams['font.size'] = 16
plt.rcParams['lines.markersize'] = 14
plt.rcParams['lines.linewidth'] = 2
```

```python
data = pd.read_csv('../run_scripts/scaling_results_003.txt', delim_whitespace=True)
```

```python
dummy_data = pd.read_csv('../run_scripts/scaling_results_dummy_003.txt', delim_whitespace=True)
```

```python
# Compute throughputs
train_throughput = data.n_steps * data.batch_size / data.train_time
data = data.assign(train_throughput=train_throughput)
data
```

```python
# Compute throughputs for dummy data
train_throughput = dummy_data.n_steps * dummy_data.batch_size / dummy_data.train_time
dummy_data = dummy_data.assign(train_throughput=train_throughput)
dummy_data
```

```python
plt.figure(figsize=(8, 6))

# Real and dummy data
plt.plot(data.n_workers, data.sust_flops, '.-', label='real data')
plt.plot(dummy_data.n_workers, dummy_data.sust_flops, '.-', label='dummy data')

# Ideal scaling
plt.plot(data.n_workers, data.n_workers * data.sust_flops.loc[0], '--', label='ideal')

plt.xlabel('Number of workers')
plt.ylabel('Sustained TFLOP/s [FP32]')

plt.legend(loc=0);
```

```python
plt.figure(figsize=(8, 6))

# Real and dummy data
plt.plot(data.n_workers, data.peak_flops, '.-', label='real data')
plt.plot(dummy_data.n_workers, dummy_data.peak_flops, '.-', label='dummy data')

# Ideal scaling
plt.plot(data.n_workers, data.n_workers * data.peak_flops.loc[0], '--', label='ideal')

plt.xlabel('Number of workers')
plt.ylabel('Peak TFLOP/s [FP32]')

plt.legend(loc=0);
```

```python
plt.figure(figsize=(8, 6))

# Real and dummy data
plt.plot(data.n_workers, data.train_throughput, '.-', label='real data')
plt.plot(dummy_data.n_workers, dummy_data.train_throughput, '.-', label='dummy data')

# Ideal scaling
plt.plot(data.n_workers, data.n_workers * data.train_throughput.loc[0], '--', label='ideal')

plt.xlabel('Number of workers')
plt.ylabel('Train throughput [samples/s]')
plt.legend(loc=0);
```

```python
scaling_eff = data.train_throughput / (data.n_workers * data.train_throughput.loc[0])
```

```python
data[['n_workers', 'train_throughput']].assign(eff=scaling_eff)
```
